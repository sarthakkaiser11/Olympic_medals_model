<h2>Preprocessing</h2>

```python
teams[teams.isnull().any(axis=1)]
```

```python
teams=teams.dropna()
```

```python
teams
```

```python
train=teams[teams["year"]<2012].copy()
```

```python
test=teams[teams["year"]>=2012].copy()

```

```python
train.shape
```

```python
test.shape
```

<h2>Training Model</h2>

```python
!pip install scikit-learn
```

```python
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
```

```python
predictors=["athletes","prev_medals"]
target="medals"
```

```python
reg.fit(train[predictors],train[target])
```

```python
predictions=reg.predict(test[predictors])
```

```python
predictions
```

```python
test["predictions"]=predictions
```

```python
test
```

```python
test.loc[test["predictions"]<0, "predictions"]=0
```

```python
test
```

```python
test["predictions"]=test["predictions"].round()
```

```python
test
```

<h2>Error Analysis</h2>

```python
from sklearn.metrics import mean_absolute_error
error=mean_absolute_error(test["medals"],test["predictions"])
```

```python
error
```

```python
teams.describe()["medals"]
```

```python
import matplotlib.pyplot as plt

# Assuming 'test' DataFrame is already defined and has 'medals' and 'predictions' columns

# Plotting the actual medals as scatter points
plt.scatter(test.index, test["medals"], color='r', label='Actual Medals', marker='x')

# Plotting the predicted medals as scatter points
plt.scatter(test.index, test["predictions"], color='b', label='Predicted Medals')

# Connecting the actual and predicted medals with lines
for i in range(len(test)):
    plt.plot([test.index[i], test.index[i]], [test["medals"].iloc[i], test["predictions"].iloc[i]], color='gray', linestyle='--', linewidth=0.5)

plt.xlabel("Index")
plt.ylabel("Medals")
plt.title("Actual vs Predicted Medals")
plt.legend()
plt.show()
```
